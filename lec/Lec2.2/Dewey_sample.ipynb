```python
# Get a product path: Choose your product and Get / Subscribe → Connect to API then you can get API endpoint (product path). Make a copy.
```

```python
# API Key
apikey_ = "YOUR API KEY"

# Pass_by Product path
pp_ = "YOUR PRODUCT PATH"
```

```python
files_df = ddp.get_file_list(apikey_, pp_, 
                             print_info = True);
```
